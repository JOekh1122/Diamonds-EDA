## ⏫⏫⏫⏫⏫

There is wrong data a clarity column with value 8  so drop it

There is data entered in wron format like ['ideal', 'very good','e'] so it will be fixed to ['Ideal','Very Good','E']




```python
df['color'] = df['color'].replace('e', 'E')
df['cut']=df['cut'].replace('ideal','Ideal')
df['cut']=df['cut'].replace('very good','Very Good')
df = df[df['clarity'] != '8']


cut=list(df['cut'].unique())
colors=list(df['color'].unique())
clarity=list(df['clarity'].unique())
print(cut)
print(colors)
print(clarity)
```

```python
sns.scatterplot(x='carat', y='price', data=df)
plt.title("Carat vs. Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

# **I think that there is colleration between Carat and Price so i will plot heatmap to see.**

```python
sns.heatmap(df[['carat', 'price']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Carat and Price")
plt.show()
```
